```python
# function to remove outliers
def select_outliers(df, column, threshold=3):
    mean = df[column].mean()
    std = df[column].std()
    df_outliers = df[df[column] > 10 *mean ]
    return df_outliers
```

```python
# Print outliers
df_original = pd.read_csv('sapipin_original.csv')
# df_original.describe()
column = 'RP1'
result = select_outliers(df_original, column)
print(result)
```

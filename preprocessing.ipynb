```python
# function to remove outliers
def select_outliers(df, column, threshold=3):
    mean = df[column].mean()
    std = df[column].std()
    # df_outliers = df[(df[column] < mean - threshold*std) | (df[column] > mean + threshold*std)]
    # df_outliers = df[df[column] > mean + threshold*std]
    df_outliers = df[df[column] > 10 *mean ]
    return df_outliers
    # df_clean = df[(df[column] > mean - threshold*std) & (df[column] < mean + threshold*std)]
    # return df_clean
```

```python
# TSNE
from sklearn.manifold import TSNE
from sklearn.preprocessing import LabelEncoder

def plot_tsne(df, classes = [10, 30, 40]):
    df = df.loc[df[df.columns[-1]].isin(classes)]
    rows, cols = df.shape
    X = df.values
    X = X[:, 0:cols]
    y = X[:, -1]
    print(y.shape)
    print(y)
    y = LabelEncoder().fit_transform(y)
    print(y)
    tsne = TSNE(n_components=2, init='pca', random_state=42)
    X_2d = tsne.fit_transform(X)

    target_ids = np.unique(y)
    for i in target_ids:
        plt.scatter(X_2d[y == i, 0], X_2d[y == i, 1])
    legendstr =[]
    for i in classes:
        legendstr.append("USER "+str(i))
    plt.legend( legendstr)
    # plt.savefig( output_name)
    plt.show()

plot_tsne(df)
```

```python
# compute a new dataset --> AVG_HT, AVH_RP, userid
# plot the new dataset
y = set(df['user'])
df1 = pd.DataFrame(columns = ['AVG_HT', 'AVG_RP', 'user'])
for user in y:
    userset = [user]
    df_user = df.loc[df[df.columns[-1]].isin(userset)]
    means = df_user.mean()
    df1 = df1.append({'AVG_HT': means['AVG_HT'], 'AVG_RP': means['AVG_RP'], 'user': user}, ignore_index = True)
df1.describe()
X = df1.iloc[:,:-1]
y = df1['user']

print(df1.shape)
# df1.to_csv("sapipin_to_clustering_111.csv")
```

```python
from sklearn.cluster import KMeans

# kmeans = KMeans(n_clusters=2, random_state=0) 

# kmeans.fit(X)

cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

```python
from sklearn.cluster import KMeans;
cluster = KMeans(3);
cluster.fit(X);

colors = ['r', 'b', 'g', 'y']
c=[colors[i] for i in cluster.labels_]
plt.scatter(X['AVG_HT'],X['AVG_RP'], color = c)  
plt.show()
```
